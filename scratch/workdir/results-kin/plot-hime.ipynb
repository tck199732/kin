```python
%load_ext autoreload
%autoreload 2
import os
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
from pathlib import Path

MPL_CUSTOM = {
    'font.family': 'serif',
    'mathtext.fontset': 'cm',
    'figure.dpi': 120,
    'figure.figsize': (4, 3.5),
    'figure.facecolor': 'white',
    'xtick.top': True,
    'xtick.direction': 'in',
    'xtick.minor.visible': True,
    'ytick.right': True,
    'ytick.direction': 'in',
    'ytick.minor.visible': True,
}

for k, v in MPL_CUSTOM.items():
    mpl.rcParams[k] = v
```

```python
pth = 'hime-hit-pattern.txt'
hist = np.loadtxt(pth)
```

```python
hist.shape
```

```python
x = np.linspace(-1000, 1000, 1001)
x = 0.5 * (x[:-1] + x[1:])
xx, yy = np.meshgrid(x, x, indexing='ij')

fig, ax = plt.subplots()
ax.hist2d(
    xx.flatten(), yy.flatten(), 
    bins=(500,500), 
    range=((-1000, 1000), (-1000, 1000)),
    weights=hist.flatten(), 
    cmap='jet', 
    norm=mpl.colors.LogNorm()
)
plt.show()
```
