```python
%load_ext autoreload
%autoreload 2
import os
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
from pathlib import Path

MPL_CUSTOM = {
    'font.family': 'serif',
    'mathtext.fontset': 'cm',
    'figure.dpi': 120,
    'figure.figsize': (4, 3.5),
    'figure.facecolor': 'white',
    'xtick.top': True,
    'xtick.direction': 'in',
    'xtick.minor.visible': True,
    'ytick.right': True,
    'ytick.direction': 'in',
    'ytick.minor.visible': True,
}

for k, v in MPL_CUSTOM.items():
    mpl.rcParams[k] = v
```

```python
hime_scaler_diff = np.loadtxt(
    'himeScalerDiff.txt'
)

samurai_scaler_diff = np.loadtxt(   
    'samuraiScalerDiff.txt'
)
```

```python
df_hime = pd.DataFrame({
    'x' : range(1, len(hime_scaler_diff) + 1),
    'y': hime_scaler_diff
})

df_samurai = pd.DataFrame({
    'x' : range(1, len(samurai_scaler_diff) + 1),
    'y': samurai_scaler_diff
})
```

```python
df_hime.head()
```

```python
df_hime_plt = df_hime.query('y <= 1e9')
df_samurai_plt = df_samurai.query('y <= 1e9')
```

```python
fig, ax = plt.subplots(dpi=300, constrained_layout=True)
ax.plot(
    df_hime_plt['x'], 
    df_hime_plt['y'], 
    color='limegreen',
    label='HIME',
    alpha=0.8
)

ax.plot(
    df_samurai_plt['x'], 
    df_samurai_plt['y'], 
    color='royalblue',
    label='SAMURAI',
    alpha=0.4
)

ax.legend() 
ax.set_xlabel('event ID')
ax.set_ylabel('diff. in time stamp')

# ax.set_xlim(25000,26000)
# ax.set_xlim(100,500)
# ax.set_xlim(100,500)

```
